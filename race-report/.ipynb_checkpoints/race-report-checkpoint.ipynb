# Frequency as a function of time ($\frac{cycles}{minute}$)

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import NamedTuple, Tuple, List
from IPython.core.display import display, HTML
import seaborn as sns

class Stats(NamedTuple):
    std: float
    mean: float
```

```python
[COL_MES, COL_TIME, COL_Y, COL_X, COL_V, VAL_CYCLE, VAL_BO] = [
    "measurement",
    "interval-time",
    "frequency",
    "distance",
    "speed",
    "cycle",
    "breakout"]
# 25 for SCM, 50 for LCM.
LAP_LEN = 25

fig, ax = plt.subplots(nrows=3, figsize=(13, 17))


def __calc_freq(df: pd.DataFrame) -> Tuple[Stats, pd.DataFrame]:
    """
    Adds the frequency column to the given dataframe, and
    calculates distribution statistics.
    """
    df_tmp = df.copy()
    frequencies = df_tmp[COL_Y]
    mean, std = np.mean(frequencies), np.std(frequencies)
    lower, upper = mean-std, mean+std
    df_tmp[COL_Y] = df_tmp[COL_Y].clip(lower=lower, upper=upper).round(2)
    df_tmp[COL_X] = df_tmp.index
    return (Stats(std=std, mean=mean), df_tmp)


def calc_speed(df: pd.DataFrame, index: List[int]) -> pd.DataFrame:
    """
    Calculates the speed per distance checkpoint (descrete speed as a function of distance).
    """
    def __calc_diff(row: pd.Series) -> float:
        return row.iloc[1] - row.iloc[0] if len(row) > 1 else row.iloc[0]

    (col_d_split, col_t_split) = ("split-distnace", "split-time")
    df_tmp = df.copy()
    (rel_col_d, rel_col_t) = (
        df_tmp[col_d_split] \
            .rolling(window=2, min_periods=1) \
            .apply(__calc_diff),
        df_tmp[col_t_split] \
        .rolling(window=2, min_periods=1) \
        .apply(__calc_diff))
    df_tmp[COL_V] = rel_col_d / rel_col_t
    return  df_tmp[~np.isnan(df_tmp[col_d_split])].rename(columns={col_d_split: COL_X})


def __add_vertical(graph:plt.figure, idx: int) -> None:
    graph.axvline(idx, linestyle=":", color="green")

    
def __get_bo_indices(df: pd.DataFrame) -> List[int]:
    ret = [idx for idx in df[df[COL_MES] == VAL_BO].index[1:]]
    ret.append(df.index[-1])
    return ret


def __format_distance(d: int) -> str:
    """
    Formats a human readable label for the given distance.
    """
    return f"{d}m"

# Distance ticks/labels
lap_indices = __get_bo_indices(df)
distances = [__format_distance((i + 1)*LAP_LEN) for i in range(len(lap_indices))]

# Frequency Dataframe.
df_frequency = df[[COL_MES, COL_TIME]][df[COL_MES] == VAL_CYCLE]
df_frequency.loc[:, COL_Y] = 60 / df[COL_TIME]

# Speed Dataframe.
df_speed = calc_speed(df, lap_indices)

# DPS.
(stats, df_frequency) = __calc_freq(df_frequency)
display(df_frequency.drop(["interval-time", COL_X, COL_MES], axis=1) \
    .style.set_caption("Frequency Data") \
    .format('{:,.1f}'))
display(pd.DataFrame({"std": [stats.std],
                      "mean": [stats.mean]}) \
    .style.set_caption('Frequency Statistics') \
    .format('{:,.1f}'))

# Plotting.

#
# Distances
#
graph = sns.lineplot(data=df_frequency, y=COL_Y, x=COL_X, ax=ax[0])
for lap_idx in lap_indices:
    __add_vertical(graph, lap_idx)
ax[0].set_xticks(labels=distances, ticks=lap_indices)
# TODO: should the limit be dynamic?
ax[0].set_ylim([30, 70])
ax[0].set_title("Frequency/Distance")

#
# Speed
#
graph = sns.lineplot(data=df_speed, y=COL_V, x=COL_X, ax=ax[1])
col_distance = df_speed[COL_X]
for d in col_distance:
    __add_vertical(graph, d)
ax[1].set_xticks(labels=[__format_distance(d) for d in col_distance], ticks= df_speed[COL_X])
# TODO: should the limit be dynamic?
ax[1].set_ylim([1.0, 3.0])
ax[1].set_title("Speed/Distance")

#
# DPS
#
num_cycles = []
for m in df[COL_MES]:
    if m == VAL_BO:
        num_cycles.append(0)
    else:
        num_cycles[-1] += 1
df_stroke_count = pd.DataFrame({"Lap": range(1, len(num_cycles)+1), "#Cycles": num_cycles})
sns.barplot(data=df_stroke_count, y="#Cycles", x="Lap", alpha=0.3, ax=ax[2])
ax[2].set_title("#Cycles/Lap")
plt.show()
```
